```python
import pandas as pd
import numpy as np
import re
import ast
```

```python
data_frame = pd.read_csv("movie data.csv")
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.isnull().sum()
```

```python
#removing rows with missing title
data_frame = data_frame.dropna(subset=["title"])


# Filling missing rate_type with 'Unknown'
data_frame["rate_type"] = data_frame["rate_type"].fillna("Unknown")
```

```python
data_frame.head()
```

```python
(data_frame[["reviews"]].head(50))
```

```python
#replacing reviews and replace string datatype by float datatype
data_frame["reviews_num"] = data_frame["user review num"].str.replace(" reviews","").str.replace(",","").astype(float)
data_frame = data_frame.drop(columns=["user review num"])
```

```python
# Convert genre from string to list
data_frame["genre"] = data_frame["genre"].apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else [])
# Explode the 'genre' list into separate rows,this will create a new row for each genre
#data_frame = data_frame.explode("genre")


```

```python
# rated_by 3.1M or 125K into numeric value

ext = data_frame["rated_by"].astype(str).str.extract(r"\(?\s*([0-9]*\.?[0-9]+)\s*([MK]?)\s*\)?", expand=True)

# Convert first group to numeric
num = pd.to_numeric(ext[0], errors="coerce")

mult = ext[1].map({"M": 1_000_000, "K": 1_000}).fillna(1).astype(float)

data_frame["rated_by"] = (num * mult).astype("float")

```

```python

# Keeping track of truly null rows
raw = data_frame["Box Office Revenue"]
null_mask = raw.isna()

# turn literal "\n" into real newlines, strip the leading "['" and trailing "']"
s = raw.astype(str)
s = s.str.replace("\\n", "\n", regex=False)
s = s.str.replace(r"^\['", "", regex=True).str.replace(r"'\]$", "", regex=True)
# Treat literal "[]" / "nan" as missing
s = s.replace({"[]": np.nan, "nan": np.nan})

#  Extract each metric will be NaN if the label isn't present
budget = s.str.extract(r"Budget\s*\$([\d,]+)", expand=False)
us_canada = s.str.extract(r"Gross US\s*&\s*Canada\s*\$([\d,]+)", expand=False)
opening = s.str.extract(r"Opening weekend US\s*&\s*Canada\s*\$([\d,]+)", expand=False)
worldwide = s.str.extract(r"Gross worldwide\s*\$([\d,]+)", expand=False)

# Clean commas and convert to numeric
data_frame["budget"] = pd.to_numeric(budget.str.replace(",", "", regex=False), errors="coerce")
data_frame["gross_us_canada"] = pd.to_numeric(us_canada.str.replace(",", "", regex=False), errors="coerce")
data_frame["opening_weekend_us_canada"] = pd.to_numeric(opening.str.replace(",", "", regex=False), errors="coerce")
data_frame["worldwide_revenue"] = pd.to_numeric(worldwide.str.replace(",", "", regex=False), errors="coerce")

# Ensure rows that were truly null stay NaN (not zero)
data_frame.loc[null_mask, ["budget","gross_us_canada","opening_weekend_us_canada","worldwide_revenue"]] = np.nan


data_frame = data_frame.drop(columns=["Box Office Revenue"])
```

```python
data_frame["release_raw"] = data_frame["release_year"].astype(str)

# Series = rows where the raw string contains a dash "–" or "-"
data_frame["type"] = "Movie"
data_frame.loc[data_frame["release_raw"].str.contains("–|-", regex=True, na=False), "type"] = "Series"

# release_year to be numeric first year only
data_frame["release_year"] = data_frame["release_raw"].str.extract(r"(\d{4})").astype("Int64")


print(data_frame[["title","release_raw","release_year","type"]].head(15))
```

```python
data_frame.to_csv("movies_after_clean1.csv", index=False)
```

```python
import pandas as pd
import ast

df = pd.read_csv("movie data.csv")

def safe_literal_list(s):
    try:
        return ast.literal_eval(s)
    except:
        return []
```

```python
rows = []

for _, row in df.iterrows():
    movie = row['title']
    reviews_list = safe_literal_list(row['reviews'])

    for review in reviews_list:
        rows.append({
            'movie_title': movie,
            'review_rating_raw': review.get('rating'),
            'review_title': review.get('title'),
            'review_text': review.get('body')
        })

reviews_long = pd.DataFrame(rows)
```

```python
reviews_long.to_csv("reviews_long.csv", index=False)
print("Extracted reviews saved to reviews_long.csv")
```

```python
import pandas as pd

df = pd.read_csv("reviews_long.csv")

def label_from_rating(r):
    if pd.isna(r): return "Unknown"
    if r >= 8: return "Positive"
    elif r <= 4: return "Negative"
    else: return "Neutral"

df['sentiment'] = df['review_rating_raw'].apply(label_from_rating)

```

```python
#removing rows with missing title
df = df.dropna(subset=["movie_title"])
```
